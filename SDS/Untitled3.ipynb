```python
import matplotlib.pyplot as plt
import numpy as np

x = [1,2,3,4]
height=[2,4,5,3]
labels=['Cars','Bikes','Bicycles','Pedestratians']

y1=np.arange(1,10)
plt.bar(x,height,align='center')
plt.xticks(x,labels)
plt.yticks(y1)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = [1,2,3,4]
height=[2,4,5,3]
labels=['Cars','Bikes','Bicycles','Pedestratians']

y1=np.arange(1,10)
plt.barh(x,height,align='center')
plt.xticks(y1)
plt.yticks(x,labels)

plt.show()
```

```python
x = [1,2,3,4]
men = [4,3,8,5]
women = [2,5,9,10]
labels = ['athletics','aquatics','hockey','tennis']
p1 = plt.bar(x,men)
p2 = plt.bar(x,women,bottom = men)
plt.xticks(x,labels)
plt.legend((p1,p2),('Men','Women'))
plt.show()
```

```python
x = np.arange(5) 
y1 = [34,56,12,89,67]
y2 = [12,56,78,45,90]
width = 0.40

plt.bar(x-0.2,y1,width)
plt.bar(x+0.2,y2,width)
# plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

N=500
x=np.random.rand(N)
y=np.random.rand(N)
colors=('r')
area=np.pi*3

plt.scatter(x,y,s=area, c=colors,alpha=0.5)

plt.title('Scatter plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
